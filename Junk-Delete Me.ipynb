```python
import numba
import numpy as np
import matplotlib
```

```python
arr = np.random.randint(90000)
```

```python
def sum_arr(arr):
    total = 0
    for i in range(300):
        total += arr[i]
    return total
```

```python
from numba import jit
```

```python
%timeit sum_arr(arr)
```

```python
print(sum_arr(arr))
```

```python
sum_arr_numba = jit()(sum_arr(arr))
```
